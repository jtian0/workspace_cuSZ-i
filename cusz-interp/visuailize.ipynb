```python
import numpy as np
import torch as th
import matplotlib.pyplot as plt

# 假设数据在一个二进制文件中，其数据类型为float32，大小为1800*3600
# 如果您已经有data作为一个numpy数组，可以跳过fromfile部分，直接使用data。
file_name = 'ODV_bcar1_1_1800_3600.dat'
data_file = [
	# '/pscratch/sd/s/swu264/SC24_ARTIFACT/data/SDRBENCH-EXASKY-NYX-512x512x512/baryon_density.f32',
	# '/pscratch/sd/s/swu264/SC24_ARTIFACT/data/SDRBENCH-EXASKY-NYX-512x512x512/baryon_density.f32.cuszx',
	f"/pscratch/sd/s/swu264/sdrbench/SDRBENCH-CESM-ATM-cleared-1800x3600/{file_name}",
	f"/pscratch/sd/s/swu264/sdrbench/SDRBENCH-CESM-ATM-cleared-1800x3600/{file_name}.cuszx",
	f"/pscratch/sd/s/swu264/sdrbench/SDRBENCH-CESM-ATM-cleared-1800x3600/{file_name}.cuszpx",
  ]  # 替换为您的数据文件名

def visualize(data_file):
	data = np.fromfile(data_file, dtype=np.float32).reshape((1800, 3600))
	# data = np.fromfile(data_file, dtype=np.float32).reshape((512, 512, 512))
	data = th.from_numpy(data)
	# 创建一个绘图窗口并设置尺寸（可根据需要调整）
	print(data.sum(dim=0, keepdim=True)[0].shape)
	plt.figure(figsize=(12, 6))

	# 使用imshow显示数据
	# origin='lower' 设置原点在左下角，cmap可以根据需要选择合适的colormap，如'viridis'、'jet'等
	plt.imshow(data, cmap='viridis', origin='lower')
	# plt.imshow(data.sum(dim=0, keepdim=True)[0], cmap='viridis', origin='lower')

	# 添加颜色条，帮助解释数值
	plt.colorbar(label='Data values')

	# 添加标题和坐标轴标签
	plt.title("1800x3600 Float Image Visualization")
	plt.xlabel("X index")
	plt.ylabel("Y index")

	# 显示图像
	plt.show()
```
